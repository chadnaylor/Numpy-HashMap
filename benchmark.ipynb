```python
from hashmap import HashMap
from hashmap_list import HashMapList
from time import process_time
import matplotlib.pyplot as plt
```

```python
times = {'get': {'hashmap': [], 'hashmap_list':[], 'dict': [], 'samples': []}, 'add': {'hashmap': [], 'hashmap_list':[], 'dict': [], 'samples': []}, 'contains': {'hashmap': [], 'hashmap_list':[], 'dict': [], 'samples': []}, 'drop': {'hashmap': [], 'hashmap_list':[], 'dict': [], 'samples': []}, 'len': {'hashmap': [], 'hashmap_list':[], 'dict': [], 'samples': []}}

for x in range(10):
    samples = 2**(11+x)
    test_dict = {i: f"val {i}" for i in range(samples)}
    hashmap = HashMap(test_dict)
    hashmap_list = HashMapList(test_dict)


    start = process_time()
    for i in range(samples):
        hashmap[i] 
    end = process_time()
    hashmap_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        hashmap_list[i] 
    end = process_time()
    hashmap_list_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        test_dict[i] 
    end = process_time()
    dict_time = round((end-start)/samples, 12)

    res = {'hashmap': hashmap_time,'hashmap_list': hashmap_list_time, 'dict': dict_time, 'samples': samples}

    for key in res:
        times['get'][key].append(res[key])

    start = process_time()
    for i in range(samples):
        del hashmap[i] 
    end = process_time()
    hashmap_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        del hashmap_list[i] 
    end = process_time()
    hashmap_list_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        del test_dict[i] 
    end = process_time()
    dict_time = round((end-start)/samples, 12)

    res = {'hashmap': hashmap_time,'hashmap_list': hashmap_list_time, 'dict': dict_time, 'samples': samples}

    for key in res:
        times['drop'][key].append(res[key])

    start = process_time()
    for i in range(samples):
        hashmap[-i] = i
    end = process_time()
    hashmap_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        hashmap_list[-i] = i
    end = process_time()
    hashmap_list_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        test_dict[-i] = i
    end = process_time()
    dict_time = round((end-start)/samples, 12)

    res = {'hashmap': hashmap_time,'hashmap_list': hashmap_list_time, 'dict': dict_time, 'samples': samples}

    for key in res:
        times['add'][key].append(res[key])

    start = process_time()
    for i in range(samples):
        i in hashmap
    end = process_time()
    hashmap_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        i in hashmap_list
    end = process_time()
    hashmap_list_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        i in test_dict 
    end = process_time()
    dict_time = round((end-start)/samples, 12)

    res = {'hashmap': hashmap_time,'hashmap_list': hashmap_list_time, 'dict': dict_time, 'samples': samples}


    for key in res:
        times['contains'][key].append(res[key])

    start = process_time()
    for i in range(samples):
        len(hashmap)
    end = process_time()
    hashmap_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        len(hashmap_list)
    end = process_time()
    hashmap_list_time = round((end-start)/samples, 12)

    start = process_time()
    for i in range(samples):
        len(test_dict) 
    end = process_time()
    dict_time = round((end-start)/samples, 12)

    res = {'hashmap': hashmap_time,'hashmap_list': hashmap_list_time, 'dict': dict_time, 'samples': samples}

    for key in res:
        times['len'][key].append(res[key])
```

```python
data = times['get']
# plt.axes(ylim= {'bottom': 100, 'ymin':0.0})
x1 = data['samples']
y1 = data['dict']
# plotting the line 1 points 
plt.plot(x1, y1, label = "dict")
# line 2 points
x2 = x1
y2 = data['hashmap']
# plotting the line 2 points 
plt.plot(x2, y2, label = "HashMap")


x2 = x1
y2 = data['hashmap_list']
# plotting the line 2 points 
plt.plot(x2, y2, label = "HashMapList")

plt.plot([0], [0])

plt.xlabel('Number of items')
# Set the y axis label of the current axis.
plt.ylabel('Average Time')
# Set a title of the current axes.
plt.title('Get times comparison')
# show a legend on the plot
plt.legend()


# Display a figure.
plt.show()
```
